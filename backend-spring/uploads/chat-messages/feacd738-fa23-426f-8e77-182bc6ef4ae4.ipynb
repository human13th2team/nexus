```python
while True:
    a = input("첫번째 숫자를 입력하세요 (종료:q): ")
    if a == "q":
        print("프로그램을 종료합니다")
        break

    b = input("두번째 숫자를 입력하세요 (종료:q): ")
    if b == "q":
        print("프로그램을 종료합니다")
        break

    c = input("연산자를 입력하세요: ")

    a = int(a)
    b = int(b)

    if c == "/" and b == 0:
        print("0으로 나눌 수 없습니다. 다시 입력하세요.")
        continue

    if c == "+":
        result = a + b
    elif c == "-":
        result = a - b
    elif c == "*":
        result = a * b
    elif c == "/":
        result = a / b
    else:
        print("잘못된 연산자 입니다.")
        continue

    print("결과:", result)
```

```python
def check_number(num):
    """숫자가 양수, 음수, 또는 0인지 확인"""
    if num > 0:
        return "양수입니다."
    elif num < 0:
        return "음수입니다."
    else:
        return "0입니다."

print(check_number(5))
print(check_number(-3))
print(check_number(0))
```

```python
def login_check(username, password):
  """ 사용자 이름과 비밀번호를 확인"""
  if username == "admin" and password == "1234":
    return "로그인 성공"
  else:
    return "로그인 실패"

print(login_check("admin", "1234"))
print(login_check("admin", "5678"))
```

```python
def find_max(numbers):
  max_num = numbers[0]
  for num in numbers:
    if num > max_num:
      max_num = num
  return max_num

numbers = [1, 5, 3, 9, 2]
print(find_max(numbers))
```

```python
def get_even_numbers(start, end):
  even_nums = []
  for num in range(start, end+1):
    if num % 2 == 0:
      even_nums.append(num)
  return even_nums

print(get_even_numbers(1, 10))
```

```python
def saving_simulation(target_amount, monthly_saving):
  total = 0
  months = 0
  while total < target_amount:
    total += monthly_saving
    months += 1
    print(f"{months}개월 차: {total: ,}원 저축")
  return f"{months}개월 만에 목표 금액 달성"

print(saving_simulation(1000000, 200000))
```

```python
def guess_number(answer):
  count = 0
  guess = None
  while guess != answer:
    guess = int(input("1부터 100 사이의 숫자를 맞춰보세요:"))
    count += 1
    if guess < answer:
      print("더 큰 숫자입니다!")
    elif guess > answer:
      print("더 작은 숫자입니다!")
  return f"정답입니다! {count}번 만에 맞추셨습니다."

print(guess_number(45))
```

```python
def normalize_data(numbers):

  min_value = min(numbers)   
  max_value = max(numbers)  
  normalized = []   

  for num in numbers:
  
    norm_value = (num - min_value) / (max_value - min_value)
    normalized.append(norm_value)
  return normalized

data = [10, 20, 30, 40, 50]
print("===== 데이터 정규화 예제 =====")
print("원본 데이터:", data)
normalized = normalize_data(data)
print("정규화된 데이터:", normalized)
```

```python
def calculate(num1, num2, operator):
    if operator == "/" and num2 == 0:
        return "Division by zero"
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        return num1 / num2
    else:
        return "Invalid operator"


# 🔥 입력은 밖에서
num1 = int(input("첫번째 숫자를 입력하세요: "))
num2 = int(input("두번째 숫자를 입력하세요: "))
operator = input("연산자를 입력하세요: ")

# 🔥 함수 호출
result = calculate(num1, num2, operator)

print("결과:", result)
```
